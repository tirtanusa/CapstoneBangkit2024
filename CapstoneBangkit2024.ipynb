# EDA

## What is the most popular genres?

```python
df.describe(include='all')
```

Using One Hot Encoding For genres

```python
from sklearn.preprocessing import MultiLabelBinarizer

def split_genres(genres):
    splitted_genres = genres.split(',')
    
    for i in range(len(splitted_genres)):
        splitted_genres[i] = splitted_genres[i].strip()
    return splitted_genres

df['genres'] = df['genres'].apply(lambda x: x.lower()) 
df['genres'] = df['genres'].apply(lambda x: split_genres(x)) #splitting each genres

mlb = MultiLabelBinarizer()

genre_encoded = mlb.fit_transform(df['genres'])

df_genres_encoded = pd.DataFrame(genre_encoded, columns=mlb.classes_)

one_hot_genres = pd.concat([df['title'], df_genres_encoded], axis =1)

print(one_hot_genres.columns)
```

```python
one_hot_genres.head()
```

```python
import matplotlib.pyplot as plt

# Get the list of genre columns
list_of_columns = one_hot_genres.columns[1:]

# Calculate the amount of films for each genre
amount = [one_hot_genres[column].sum() for column in list_of_columns]

print(amount, list_of_columns)

# Zip genre names and their corresponding amounts, then sort by amount
sorted_genres = [x for _, x in sorted(zip(amount, list_of_columns), reverse=True)]

# Sort the amounts in descending order
sorted_amount = sorted(amount, reverse=True)

#Plotting
plt.figure(figsize=(10, 6))

# Plot the bar for the category with the highest amount of films in red
plt.bar(sorted_genres[0], sorted_amount[0], color='red', label=sorted_genres[0])

# Plot the bars for the other categories in light grey
for category, film in zip(sorted_genres[1:], sorted_amount[1:]):
    plt.bar(category, film, color='lightgrey')

plt.title('Most Popular Genre')
plt.xlabel('Category')
plt.ylabel('Films')

plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

## What is the most profitable Movies

```python
df['profit'] = df['revenue'] - df['budget']

df
```

```python
profitable_movies = df.sort_values(by='profit', ascending=False)
profitable_movies.drop(index=[212755, 336036], inplace=True)
profitable_movies.head(10)
```

```python
# Select the top 10 most profitable movies
top_10_profitable_movies = profitable_movies.head(10)

# Plotting
plt.figure(figsize=(10, 6))


# Plotting the first movie in red
plt.bar(top_10_profitable_movies['title'].iloc[0], 
        top_10_profitable_movies['profit'].iloc[0], 
        color='red', 
        label=top_10_profitable_movies['title'].iloc[0])

# Plotting the rest of the movies in grey
for idx in range(1, len(top_10_profitable_movies)):
    plt.bar(top_10_profitable_movies['title'].iloc[idx], 
            top_10_profitable_movies['profit'].iloc[idx], 
            color='lightgrey')

# Adding titles and labels
plt.title('Top 10 Most Profitable Movies')
plt.xlabel('Movie Title')
plt.ylabel('Profit')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()
plt.show()
```

## Most Profitable Studio/Companies

```python
df['production_companies'] = df['production_companies'].apply(lambda x: x.lower()) 
df['production_companies'] = df['production_companies'].apply(lambda x: split_genres(x)) #splitting each companies

df['production_companies']
```

```python
# Expand the DataFrame so that each row represents a single production company associated with a movie
expanded_df = df.explode('production_companies')

# Calculate profit for each production company
company_profit = expanded_df.groupby('production_companies')['profit'].sum()

# Reset index to make 'production_companies' a regular column
company_profit = company_profit.reset_index()

# Sort the DataFrame by profit in descending order
company_profit = company_profit.sort_values(by='profit', ascending=False)

# Display the DataFrame with profit for each production company
company_profit.head(10)
```
